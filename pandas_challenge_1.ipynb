```python
# Import Dependencies
import pandas as pd
```

```python
# import the data from the .csv file 
client_data = 'Resources/client_dataset.csv'
```

```python
# Read our data file with the Pandas library by creating a new dataframe
client_df=pd.read_csv(client_data)
client_df.info()
```

```python
# View the records of the data
## Data contains 54,639 rows and 17 columns
client_df
client_df.head(10)
```

```python
#Use the describe function to gather some basic statistics
client_df.describe()
```

```python
#What three item categories had the most entries?
## find how many entries each category has
category_count= client_df["category"].value_counts()
category_count
```

```python
## Print out the top 3 categories with the most entries
category_count.head(3)
```

```python
#For the category with the most entries, which subcategory had the most entries
## check the total count of sub categories
subcategory_count= client_df["subcategory"].value_counts()
subcategory_count
```

```python
# Creating a new data frame to hold the consumables sub category  and its values
consumables_subcategory = client_df.loc[client_df['category']=='consumables','subcategory'].value_counts()
consumables_subcategory
# Printing out the subcategory with the most entries
consumables_subcategory.head(1)
```

```python
#Which five clients had the most entries in the data?
##Combining fucntions to create the top 5 clients based on entries in the data
clients_usage =client_df['client_id'].value_counts().head(5)
print(f" The top 5 clients with the most entries in the data are")
clients_usage
```

```python
#How many total units (the qty column) did the client with the most entries order?
# check how many entris the client_id =33615 has
max_client_units = client_df.loc[client_df['client_id']== 33615,'qty'].sum()
client_id_max= clients_usage.index[0]
print(f"The top client {client_id_max} has {max_client_units} units")
```

```python
#Create a column that calculates the subtotal for each line using the unit_price and the qty.
## Creating the column 'subtotal that equals the unit price* the qty of each entry
client_df['subtotal']= client_df['unit_price']* client_df['qty']
client_df
```
