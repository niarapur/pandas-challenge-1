```python
# Cleaning the dataset to remove records where the unit weight =0 or quantity=0 as it is assumed client must have these two fields populated
' the below step removed 142 rows where either the unit weight or the quantity is 0'
client_df_clean = client_df.loc[(client_df['unit_weight']>0) & (client_df['qty']>0)]
client_df_clean
```

```python
#Which five clients had the most entries in the data?
##Combining fucntions to create the top 5 clients based on entries in the data
clients_usage =client_df_clean['client_id'].value_counts().head(5)
print(f" The top 5 clients with the most entries in the data are")
clients_usage
```

```python
#How many total units (the qty column) did the client with the most entries order?
# check how many entris the client_id =33615 has
max_client_units = client_df_clean.loc[client_df_clean['client_id']== 33615,'qty'].sum()
client_id_max= clients_usage.index[0]
print(f"The top client {client_id_max} has {max_client_units} units ordered")
```

```python
#Create a column that calculates the subtotal for each line using the unit_price and the qty.
## Creating the column 'subtotal that equals the unit price* the qty of each entry
"got a'SettingwithCopy Warning error here and had to correct it using .loc to define the new variable"

client_df_clean.loc[:,'subtotal'] = client_df_clean['unit_price']* client_df_clean['qty']
#client_df_clean['subtotal']= client_df_clean['subtotal'].map('{:,.2f}'.format)
client_df_clean.sort_values('subtotal')
```

```python
#Create a column for shipping price. Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
' Additional Assumptions'
' Assumming the unit_weight is in pounds'

# define a function to caclulate shipping price based on weight

def shipping_price(weight):
    if weight >50:
        shipping_price=7*weight
    else:
        shipping_price=10*weight
    return shipping_price

# Creating a total weight variable based on the order's unit weight and quantity
client_df_clean.loc[:,'total_weight'] = client_df_clean['unit_weight']*client_df_clean['qty']

# applying the function to calculate the shipping price based on the total weight
client_df_clean.loc[:,'shipping_price'] = client_df_clean['total_weight'].apply(shipping_price)
client_df_clean.sort_values(by ='total_weight', ascending=False)
#client_df_clean.sort_values(by ='total_weight')
    
```

```python
#Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%.
' Sales tax applies to both total price and the shipping price'
client_df_clean.loc[:,"total_price_without_st"] =(client_df_clean['subtotal']) + (client_df_clean['shipping_price'])
client_df_clean.loc[:,"total_price_with_st"] =(1 + 0.0925)*((client_df_clean['subtotal'])+ (client_df_clean['shipping_price']))
client_df_clean
```

```python
#Create a column for the cost of each line using unit cost, qty, and shipping price 
' Assumption: Shipping cost is the same as what is charged to the client'
client_df_clean.loc[:,"total_cost"] = (client_df_clean["unit_cost"]*client_df_clean["qty"]) + (client_df_clean["shipping_price"])
client_df_clean
```

```python
#Create a column for the profit of each line using line cost and line price.
' Assumes Sales tax is only applied to the customer and cannot be included in the profit '
client_df_clean.loc[:,"total_profit"] = client_df_clean["total_price_without_st"] - client_df_clean["total_cost"]
client_df_clean
```
